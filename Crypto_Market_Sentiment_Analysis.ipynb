```python
historical_data["Timestamp IST"] = pd.to_datetime(
    historical_data["Timestamp IST"],
    dayfirst=True
)
```

```python
historical_data["date"] = historical_data["Timestamp IST"].dt.date
```

```python
historical_data["date"] = pd.to_datetime(historical_data["date"])
```

```python
merged_df = pd.merge(
    historical_data,
    fear_greed,
    on="date",
    how="inner"
)
```

```python
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df["classification"].value_counts()
```

```python
import matplotlib.pyplot as plt

merged_df["classification"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Distribution of Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Records")

plt.show()
```

```python
merged_df.groupby("classification")["Closed PnL"].mean()
```

```python
import matplotlib.pyplot as plt

merged_df.groupby("classification")["Closed PnL"].mean().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Average Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Closed PnL")

plt.show()
```

```python
merged_df.groupby("classification")["Size USD"].sum()
```

```python
import matplotlib.pyplot as plt

merged_df.groupby("classification")["Size USD"].sum().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Total Trading Volume by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Total Size USD")

plt.show()
```

```python
pd.crosstab(
    merged_df["classification"],
    merged_df["Side"]
)
```

```python
import matplotlib.pyplot as plt

pd.crosstab(
    merged_df["classification"],
    merged_df["Side"]
).plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Buy vs Sell Activity Across Market Sentiments")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")

plt.show()
```

```python
merged_df.groupby("classification")["Coin"].value_counts().groupby(level=0).head(5)
```

```python
top_coins = (
    merged_df.groupby("classification")["Coin"]
    .value_counts()
    .groupby(level=0)
    .head(5)
    .reset_index(name="count")
)
```

```python
import matplotlib.pyplot as plt

pivot_df = top_coins.pivot(
    index="Coin",
    columns="classification",
    values="count"
)

pivot_df.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Top Traded Coins Across Market Sentiments")
plt.xlabel("Coin")
plt.ylabel("Number of Trades")

plt.xticks(rotation=45)

plt.show()
```

```python
merged_df.groupby("Direction")["Closed PnL"].mean()
```

```python
import matplotlib.pyplot as plt

merged_df.groupby("Direction")["Closed PnL"].mean().plot(
    kind="barh",
    figsize=(10,6)
)

plt.title("Average Closed PnL by Trade Direction")
plt.xlabel("Average Closed PnL")
plt.ylabel("Trade Direction")

plt.show()
```

```python
numeric_cols = merged_df.select_dtypes(include="number")

numeric_cols.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = numeric_cols.corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap of Numerical Features")

plt.show()
```
